#### 1st case:  Poisson, const prior, (n,T) = (16, 2)

```python
import numpy as np
import scipy
import matplotlib as mpl
from matplotlib.pyplot import *
from numpy import *
from scipy import stats
from univariate_bayes import UnivariateBayesianInference
from scipy.stats import norm

# Imports from the module with UnivariateBayesianInference subclasses:
from poisson_binomial_cauchy import PoissonRateInference, BinomialInference,CauchyLocationInference
```

```python
r_u = 100.  # upper limit for PDF calculation and plotting

# Create a PRI instance and plot the PDF.
prior_l, prior_u = 0., 1e5
flat_pdf = 1./(prior_u - prior_l)
n, T = 50, 2
pri = PoissonRateInference(T, n, flat_pdf, r_u)
post_mean=pri.post_mean
mlike=pri.mlike

def g_mean(params):
    """
    Return the function whose expectation gives the posterior mean, i.e.,
    just return the values of the params.
    """
    params = asarray(params)
    return params

have_laplace=True
if have_laplace:
    # Laplace approx for the marg. like. and the mean:
    ampl, locn, sig, ml = pri.laplace()
    x=pri.param_grid
    y=pri.post_pdf_la
    laplace_mean = pri.laplace(g_mean)
    post_mean_l = laplace_mean[3]/ml

    # Print using string formatting:
    print('Poisson case:')
    print('Marg. like.: {:.4e} (quad), {:.4e} (Laplace)'.format(pri.mlike, ml))
    print('Posterior mean: {:4.2f} (quad), {:4.2f} (Laplace)'.format(pri.post_mean, post_mean_l))
    
    
figure(figsize=(8,5))
pri.plot(alpha=.5)
plot(x,y,'--',label='Laplace approximation',color='g')
xlabel(r'Rate (s$^{-1}$)')
ylabel('PDF (s)')
legend()
title('Poisson case')
```

```python
#Binomial:samp_cdf
n, n_trials = 8, 12

bi = BinomialInference(n, n_trials)

figure(figsize=(8,5))
bi.plot(alpha=.5)

xlabel(r'$\alpha$')
ylabel('Posterior PDF')
title('Binomial case')

samps_cdf = []
for i in range(10000):
    samps_cdf.append(bi.samp_cdf())
samps_cdf = array(samps_cdf)

bins = hist(samps_cdf, 50, density=True, color='g', alpha=.5)
```

```python
#Bionomial:samp_ar
figure(figsize=(8,5))
bi.plot(alpha=.5)

xlabel(r'$\alpha$')
ylabel('Posterior PDF')
title('Binomial case')

samps = []
for i in range(1000):
    samps.append(bi.samp_ar())
    if i==200: print('Yeah!')
    if i==400: print('Yeah!')
    if i==600: print('Yeah!')
    if i==800: print('Yeah!')
samps = array(samps)
n_ar=bi.n_ar
bins = hist(samps, 50, density=True, color='g', alpha=.5)
```

```python
print('samp_ar:Sample Mean:',mean(samps))
print('samp_ar:Standard Deviation:',std(samps))
print('quadrature:Sample Mean:',bi.post_mean)
print('quadrature:Standard Deviation:',bi.post_std)
```

```python
#sample efficiency
print('Sample Efficiency:',len(samps)/n_ar)
```
